# Cleaning Data

```python
df["StudentID"].fillna(df["StudentID"].median(), inplace=True)
df["AttendanceRate"].fillna(df["AttendanceRate"].median(), inplace=True)
df["PreviousGrade"].fillna(df["PreviousGrade"].median(), inplace=True)
df["ExtracurricularActivities"].fillna(df["ExtracurricularActivities"].median(), inplace=True)
df["FinalGrade"].fillna(df["FinalGrade"].median(), inplace=True)
df["Attendance (%)"].fillna(df["Attendance (%)"].median(), inplace=True)
df["Name"].fillna("Unknown", inplace=True)
df["Gender"].fillna(df["Gender"].mode()[0], inplace=True)
df["StudyHoursPerWeek"].fillna(df["StudyHoursPerWeek"].mode()[0], inplace=True)
df["Online Classes Taken"].fillna(df["Online Classes Taken"].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Study Hours"]
```

```python
df["Study Hours"] = df["Study Hours"].replace(-1, np.nan)
df["Study Hours"] = df["Study Hours"].fillna(df["Study Hours"].median())
df["Study Hours"].fillna(df["Study Hours"].median(), inplace=True)
```

```python
df["ParentalSupport"].fillna("Unknown", inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Gender"] = df["Gender"].str.lower()
df["Gender"].unique()
```

# Exploratory Data Analysis

```python
print(df.groupby("Gender")["FinalGrade"].mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["Attendance (%)"], df["FinalGrade"], alpha=0.5)
plt.xlabel("SAttendance (%)")
plt.ylabel("Final Grade")
plt.title("Relationship between Attendance (%) and Final Grade")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["Study Hours"], kde=True)
plt.xlabel("Study Hours")
plt.ylabel("final grade")
plt.title("Distribution of Study Hours")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="ParentalSupport", y="FinalGrade", data=df, order=["Low", "Medium", "High"],palette="viridis")
plt.xlabel("Parental Support")
plt.ylabel("Average Final Grade")
plt.title("Average Final Grade by Parental Support Level")
plt.show()
```

```python
print(df["ExtracurricularActivities"].unique())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Online Classes Taken" ,palette="viridis")
plt.title("Final Grade Distribution by Online Classes Taken")
plt.show()
```
